```python
import math

feature_category = pd.DataFrame({'num':[], 'wt_title':[], 'distance':[], 'category':[]})
x_cate_list = [20, 0, -20]
y_cate_list = [-20, 0, 20]

for cate in range(1,10):
    for i in range(len(vec_mean_df)):
        x_devide = round(vec_mean_df['Xmean'][i]/10000, 2)
        y_devide = round(vec_mean_df['Ymean'][i]/10000, 2)
        x_len = x_devide + x_cate_list[(cate-1)%3]
        y_len = y_devide + y_cate_list[int((cate-1)/3)]
        distance = math.sqrt(math.pow(x_len,2) + math.pow(y_len,2))
        if i < 30:
            feature_category.loc[(cate-1)*30+i] = {'num':i, 'wt_title':vec_mean_df['WebToonTitle'][i], 'distance':distance, 'category':cate}
        else:
            max_value = max(feature_category['distance'][(cate-1)*30:(cate)*30])
            for j in range((cate-1)*30, (cate)*30):
                if (feature_category['distance'][j] == max_value) and (feature_category['distance'][j] > distance):
                    feature_category.loc[j] = {'num':i, 'wt_title':vec_mean_df['WebToonTitle'][i], 'distance':distance, 'category':cate}
                    break
feature_category['num'] = feature_category['num'].astype('int64')
feature_category['category'] = feature_category['category'].astype('int64')
feature_category['distance'] = feature_category['distance'].round(2)
```

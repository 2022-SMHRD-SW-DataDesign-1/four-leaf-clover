```python
from konlpy.tag import Kkma
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('data/태그별점추가.csv', encoding='utf-8')
data
```

```python
synopsis = data['synopsis'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
def myTokenizer(text) : 
    d = pd.DataFrame(kkma.pos(text),columns=['morphs','tag'])
    d.set_index('tag',inplace=True)
    if('NNG' in d.index) :
        return d.loc[d.index.intersection(['NNG']),'morphs'].values
    else :
        return []
```

```python
kkma = Kkma()

tf_idf_vect = TfidfVectorizer(tokenizer=myTokenizer)
tf_idf_matrix = tf_idf_vect.fit_transform(synopsis)
type(tf_idf_vect.vocabulary_.keys())
```

```python
from gensim.models import word2vec

model = word2vec.Word2Vec(tf_idf_vect.vocabulary_.keys(), window = 5, min_count = 1, workers = 4, sg = 1)
```

```python
chat_list=[]
chatlst=[] 
wb_list=[]
wblst=[]

wbid_list=[]
wbidlst=[]

for i in range(len(all_webtoons['webtoon_url'])):
    # 성인 웹툰이면 그냥 넘어갑시다.
    
    if all_webtoons['webtoon_url'][i] in adult_url:
        wbid_list.append(all_webtoons['id'][i])
        wb_list.append(all_webtoons['title'][i])
        chat_list.append('성인 웹툰은 로그인 필요.')
        print(all_webtoons['title'][i],"베댓 수집 완료")

        wbidlst.append(wbid_list.copy())
        wblst.append(wb_list.copy())
        chatlst.append(chat_list.copy())
        continue
    #전체 웹툰 목록 중 월요일 첫 번째 웹툰으로 페이지 이동
    driver.get(all_webtoons['webtoon_url'][i])
    time.sleep(3)
    
    # beautiful soup으로 html 정보 받고   
    html = requests.get(all_webtoons['webtoon_url'][i], headers= headers)
    result = bs(html.content,'html.parser')

    # 여기에 각 회차로 들어가는 링크도 존재함
    title = result.findAll("td", {"class", "title"})     

    # 최근 10회분의 제목을 수집 => title이라는 list의 길이가 10개 그리고 td, title 안에 a링크도같이있음
    for t in title: 
        # 각 회차별 html 소스 가져오기
        url ="https://comic.naver.com" + t.a['href'] # 각회차별 링크
        
        html2 = requests.get(url, headers = headers) # 우회 제끼고
        result2 = bs(html2.content, "html.parser") # html 정보 parser해서 옴
        driver.get(url)
        
        #현재 댓글 부분 파싱이 안되는 설정이므로 src의 댓글창 페이지로 직접 이동
        #현재 URL주소 복사
        html = driver.current_url
        currentPage=html[html.find('?'):]
        
        #댓글창 페이지 URL
        url2='https://comic.naver.com/comment/comment.nhn' + currentPage
        time.sleep(0.4)
        
        #댓글창 페이지 열기
        driver2 = webdriver.Chrome('chromedriver.exe')
        driver2.get(url2)
        
        time.sleep(5)
        
        while True:
            try:
                #수집 편의를 위해 클린봇 댓글 필터링 기능 해제
                driver2.find_element(By.CLASS_NAME,'u_cbox_cleanbot_setbutton').click()
                time.sleep(0.4)
                driver2.find_element(By.CSS_SELECTOR,'#cleanbot_dialog_checkbox_cbox_module').click()
                time.sleep(0.4)
                driver2.find_element(By.CLASS_NAME,'u_cbox_layer_cleanbot2_extrabtn').click()
                break
                
            except:
                driver2.close()
                driver2 = webdriver.Chrome('chromedriver.exe')
                driver2.get(url2)
                print('driver2 ERROR')
                time.sleep(20)
                continue
        
        
        #베댓 내용
        chatbest  = driver2.find_elements(By.CLASS_NAME,'u_cbox_contents')
        time.sleep(2)
        for k in range(len(chatbest)):
            wbid_list.append(all_webtoons['id'][i])
            wb_list.append(all_webtoons['title'][i])
            chat_list.append(driver2.find_elements(By.CLASS_NAME,'u_cbox_contents')[k].text)
            
        time.sleep(2)
        #페이지 뒤로 가기, 다시 만화 목록으로 
        driver.back()
        time.sleep(1)
        chatbest.clear()
    
    #전체 정식연재 만화 창으로 이동
    time.sleep(1)
    print(all_webtoons['title'][i],"베댓 수집 완료")
    
    wbidlst.append(wbid_list.copy())
    wblst.append(wb_list.copy())
    chatlst.append(chat_list.copy())
#---------------------------------------------------크롤링 END--------------------------------------------------------------
```

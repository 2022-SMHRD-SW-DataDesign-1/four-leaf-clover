```python
import pandas as pd
```

```python
now = pd.read_csv("data/태그별점추가.csv", encoding="utf-8")
now = now[["title","genre","synopsis","tag"]]
now
```

```python
end = pd.read_csv("data/완결웹툰정보장르추가.csv",encoding="utf-8")
end = end[["title","genre","synopsis","tag"]]
end
```

```python
total = pd.concat([now, end])
total
```

```python
total = total.replace("성인물","성인웹툰")
```

```python
total.to_csv('data/전체웹툰_제목,장르,시놉시스,태그.csv',index=False)
```

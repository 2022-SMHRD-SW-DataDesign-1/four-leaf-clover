```python
from konlpy.tag import Kkma, Twitter
import pandas as pd
```

```python
data = pd.read_csv('data/네이버 웹툰 정보ansi.csv', encoding='cp949')
data
```

```python
synopsis = data['synopsis']
synopsis
```

```python
kkma = Kkma()
twitter = Twitter()
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
mydoclist_kkma = []
mydoclist_twitter = []

for i in range(len(synopsis)) :
    kkma_nouns = ' '.join(kkma.nouns(synopsis[i]))
    twitter_nouns = ' '.join(twitter.nouns(synopsis[i]))
    mydoclist_kkma.append(kkma_nouns)
    mydoclist_twitter.append(twitter_nouns)

tf_idf_vectorizer = TfidfVectorizer(min_df = 1)
tf_idf_matrix_kkma = tf_idf_vectorizer.fit_transform(mydoclist_kkma)
tf_idf_matrix_twitter = tf_idf_vectorizer.fit_transform(mydoclist_twitter)

document_distances_kkma = (tf_idf_matrix_kkma * tf_idf_matrix_kkma.T)
document_distances_twitter = (tf_idf_matrix_twitter * tf_idf_matrix_twitter.T)

print(document_distances_kkma.toarray())
print(document_distances_twitter.toarray())
```

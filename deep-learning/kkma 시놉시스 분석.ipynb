```python
from konlpy.tag import Kkma
```

```python
import pandas as pd
```

```python
synopsis = pd.read_csv('data/제목작가시놉시스17.csv',encoding='cp949')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
synop = synopsis['story']
```

```python
synop.head()
```

```python
tf_idf_vect = TfidfVectorizer()
tf_idf_vect.fit(synop)
tf_idf_vect.vocabulary_.keys()
```

```python
kkma = Kkma()
```

```python
kkma.nouns(synop[0])
```

```python
kkma.jki(synop[0])
```

```python
def myTokenizer(text) : 
    d = pd.DataFrame(kkma.pos(text),columns=['morphs','tag'])
    d.set_index('tag',inplace=True)
    if('VV' in d.index) | ('VA' in d.index) | ('NNG' in d.index) :
        return d.loc[d.index.intersection(['VV','VA','NNG']),'morphs'].values
    else :
        return []
```

```python
tf_idf_vect = TfidfVectorizer(tokenizer=myTokenizer)
tf_idf_vect.fit(synop)
tf_idf_vect.vocabulary_
```

```python
vec_key = tf_idf_vect.vocabulary_.keys()
vec_values = tf_idf_vect.vocabulary_.values()
df = pd.DataFrame([vec_key,vec_values]).T
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr
import pandas as pd
from PIL import Image
import re
```

```python
data = pd.read_csv('./네이버 웹툰 정보ansi.csv',encoding='CP949')
```

```python
os.chdir('/Users/smhrd/Desktop/thirdproject/webtoons/')
print(os.getcwd())
```

```python
# easyOCR 리더 옵션 설정
reader = easyocr.Reader(['ko'],gpu=True)
folder_list = []
for i in range(len(data['title'])):
    folder_title = re.sub("[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]",'_', data['title'][i])
    folder_list.append(folder_title)
```

```python
### 자동화 코드
first_texts= []
each_first_texts=[]

for i in range(len(folder_list)) :
    try:
        os.chdir('./'+folder_list[i]+'/1')
        for j in range(len(os.listdir())):
            img = Image.open('./'+os.listdir()[j])
            # use numpy to convert the pil_image into a numpy array
            numpy_image=np.array(img)
            opencv_image=cv2.cvtColor(numpy_image, cv2.COLOR_RGB2BGR)
            result = reader.readtext(opencv_image)
            for k in range(len(result)):
                each_first_texts.append(result[k][1])
        first_texts.append(each_first_texts)
        each_first_texts =[]
    except:
        print(folder_list[i]+'는 성인웹툰입니다.')
        first_texts.append('성인웹툰입니다.')
        os.chdir('/Users/smhrd/Desktop/thirdproject/webtoons/')
        print(folder_list[i]+''+str(i)+'/561')
        continue
    os.chdir('/Users/smhrd/Desktop/thirdproject/webtoons/')
    print(folder_list[i]+" "+str(i)+'/561')
```

```python
from konlpy.tag import Okt
okt = Okt()
# 1. csv파일로 뽑아놓고(원본)
ocrData=pd.DataFrame()
ocrData['title']=folder_list
ocrData['ocr']=first_texts
ocrData.to_csv('제목OCR.csv',encoding='euc-kr')
```

```python
# 2. 리스트 형태로 들어 간 얘를 하나의 스트링으로 변환 ==> 이건 찍어보고
text= []
for i in range(len(ocrData['ocr'])):
    text.append(''.join(ocrData['ocr'][i]))
```

```python
# 3. 특수문자 제거 ==> 특수문자 제거 
for i in range(len(text)):
    text[i] = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z]', ' ', text[i])
```

```python
# 4.nouns_lists_list추출 (kkma)
nouns_list = []
from collections import Counter
for i in range(len(text)):
    nouns_list.append(okt.nouns(text[i]))
```

```python
# 5. 추출 명사들 중복 명사 값 제거 for 비교
test_list=[]
for i in range(len(nouns_list)):
    test_list.append(list (set(nouns_list[i])))
```
